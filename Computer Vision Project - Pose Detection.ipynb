```python
# Step 1: Import all necessary libraries
import cv2
import mediapipe as mp
```

```python
# Step 2: Identify Webcam
cap = cv2.VideoCapture(0)  #local webcam - 0, external webcam - 1
```

```python
#Step 4: Leveraging the Mediapipe library used for hand detection
mpPose = mp.solutions.pose
pose = mpPose.Pose()
# pose = mpPose(static_image_mode = False, upper_body_only = False, smooth_landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# To draw and connect the landmarks
mpDraw = mp.solutions.drawing_utils
```

```python
# Switch on webcam

while True:
    _, img = cap.read()
    
    # Convert Video/image from BGR RGB 
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Apply mediapipe pose detection module for detection
    results = pose.process(imgRGB)
    print(results.pose_landmarks)
    
    # Draw landmarks
    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
    
    cv2.putText(img, "Izzy's Pose Detection", (10, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,255), 2)
    
    cv2.imshow("Izzy's Pose Detection Project", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
